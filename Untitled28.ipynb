```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader, random_split, Subset
import os
import time
import copy
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
import sklearn.metrics as sm
import seaborn as sns # 用于绘制更美观的混淆矩阵

# **** 移除分布式训练相关模块 ****
# import torch.multiprocessing as mp
# from torch.utils.data.distributed import DistributedSampler  # 用于DDP的数据采样器
# from torch.nn.parallel import DistributedDataParallel as DDP  # DDP包装器
# from torch.distributed import init_process_group, destroy_process_group  # DDP进程组初始化和清理
# import torch.distributed as dist # 用于 all_reduce 等操作

# **** 新增混合精度训练模块 ****
from torch.cuda.amp import autocast, GradScaler

# ImageNet数据集的均值和标准差 (全局常量)
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD = [0.229, 0.224, 0.225]

# --- 1. 配置 (全局常量) ---
# 数据集路径
SPECIFIC_TRAIN_DIR = r'D:\fruits-360\Training' # 修改为D盘路径
SPECIFIC_TEST_DIR = r'D:\fruits-360\Test' # 修改为D盘路径

# InceptionV3期望输入大小为 299x299
NORMALIZE_TRANSFORM = transforms.Normalize(mean=IMAGENET_MEAN, std=IMAGENET_STD) # ImageNet归一化参数

# --- 2. 数据预处理与增强 (全局定义) ---
train_transforms = transforms.Compose([
    transforms.RandomResizedCrop(299),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    NORMALIZE_TRANSFORM
])

val_test_transforms = transforms.Compose([
    transforms.Resize(320),
    transforms.CenterCrop(299),
    transforms.ToTensor(),
    NORMALIZE_TRANSFORM
])

# --- 3. 数据加载辅助类 (全局定义) ---
class TransformedSubset(torch.utils.data.Dataset):
    def __init__(self, subset, transform):
        self.subset = subset
        self.transform = transform
    def __getitem__(self, index):
        x, y = self.subset[index]
        if self.transform:
            x = self.transform(x)
        return x, y
    def __len__(self, ):
        return len(self.subset)

# --- 4. 模型定义 (全局定义) ---
def get_model(num_classes, pretrained=True):
    # InceptionV3 默认带辅助分类器，此处确保其正确配置
    # weights 参数在 torchvision 0.13.0+ 版本中取代了 pretrained
    model = models.inception_v3(weights=models.Inception_V3_Weights.IMAGENET1K_V1 if pretrained else None, aux_logits=True)
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, num_classes)

    # 确保辅助分类器也修改为正确的类别数
    if hasattr(model, 'AuxLogits') and model.AuxLogits is not None:
        num_aux_ftrs = model.AuxLogits.fc.in_features
        model.AuxLogits.fc = nn.Linear(num_aux_ftrs, num_classes)
    return model

# --- 5. 训练和评估辅助函数 (单GPU修改，新增混合精度) ---
def train_eval_one_epoch(model, dataloader, criterion, optimizer, device, is_training, current_epoch=None, total_epochs=None, scaler=None):
    if is_training:
        model.train()
        desc_prefix = f"Epoch {current_epoch}/{total_epochs} Train"
    else:
        model.eval()
        desc_prefix = "Eval"
        if current_epoch is not None: # 用于训练期间的验证
             desc_prefix = f"Epoch {current_epoch}/{total_epochs} Valid"
        else: # 用于最终测试评估
            desc_prefix = f"Test"

    running_loss = 0.0
    running_corrects = 0
    total_samples = 0

    all_labels = []
    all_preds = []

    progress_bar_iterable = tqdm(dataloader, desc=desc_prefix, leave=False, unit="batch")

    for inputs, labels in progress_bar_iterable:
        inputs = inputs.to(device)
        labels = labels.to(device)

        if is_training:
            optimizer.zero_grad()

        with torch.set_grad_enabled(is_training):
            # **** 混合精度：使用 autocast ****
            with autocast(enabled=torch.cuda.is_available() and scaler is not None):
                outputs = model(inputs)
                
                if isinstance(outputs, tuple) and model.aux_logits and is_training:
                    main_output, aux_output = outputs
                    loss1 = criterion(main_output, labels)
                    loss2 = criterion(aux_output, labels)
                    loss = loss1 + 0.4 * loss2 # 辅助损失权重为0.4
                    preds = torch.max(main_output, 1)[1]
                else:
                    final_outputs = outputs.logits if isinstance(outputs, models.inception.InceptionOutputs) else outputs
                    loss = criterion(final_outputs, labels)
                    preds = torch.max(final_outputs, 1)[1]

            if is_training:
                # **** 混合精度：使用 scaler 进行反向传播和优化器步进 ****
                if scaler is not None:
                    scaler.scale(loss).backward()
                    scaler.step(optimizer)
                    scaler.update()
                else:
                    loss.backward()
                    optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)
        total_samples += inputs.size(0)

        if not is_training: # 仅在评估时收集标签和预测
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())

        if is_training: # 仅在训练时更新批次进度条的后缀
            batch_loss = loss.item()
            batch_acc = torch.sum(preds == labels.data).double().item() / inputs.size(0)
            progress_bar_iterable.set_postfix(loss=f"{batch_loss:.4f}", acc=f"{batch_acc:.4f}")

    epoch_loss = running_loss / total_samples if total_samples > 0 else 0.0
    epoch_acc = running_corrects.item() / total_samples if total_samples > 0 else 0.0

    if not is_training:
        return epoch_loss, epoch_acc, np.array(all_labels), np.array(all_preds)
    return epoch_loss, epoch_acc

# --- EarlyStopping 类 (单GPU修改) ---
class EarlyStopping:
    def __init__(self, patience=7, verbose=False, delta=0, path='checkpoint.pt', trace_func=print):
        self.patience = patience
        self.verbose = verbose
        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = float('inf')
        self.delta = delta
        self.path = path
        self.trace_func = trace_func

    def __call__(self, val_loss, model):
        score = -val_loss
        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model) # 单GPU直接保存
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.verbose: self.trace_func(
                f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience: self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model) # 单GPU直接保存
            self.counter = 0

    def save_checkpoint(self, val_loss, model):
        if self.verbose:
            self.trace_func(f'Validation loss decreased ({self.val_loss_min:.6f} --> {val_loss:.6f}). Saving model to {self.path}')
        torch.save(model.state_dict(), self.path)
        self.val_loss_min = val_loss

# --- 训练主循环函数 (单GPU修改，新增混合精度) ---
def train_loop(model, criterion, optimizer, dataloaders, num_epochs, early_stopping_patience, checkpoint_path, device):
    since = time.time()
    early_stopping = EarlyStopping(patience=early_stopping_patience, verbose=True, path=checkpoint_path)
    history = {'train_loss': [], 'train_acc': [], 'val_loss': [], 'val_acc': []}
    best_val_acc_observed = 0.0 # 记录训练过程中观察到的最佳验证准确率

    # **** 混合精度：初始化 GradScaler ****
    scaler = GradScaler(enabled=torch.cuda.is_available())

    for epoch in range(num_epochs):
        print(f'\nEpoch {epoch + 1}/{num_epochs}'); print('-' * 10)

        # 训练阶段
        epoch_train_loss, epoch_train_acc = train_eval_one_epoch(model, dataloaders['train'], criterion, optimizer, device, is_training=True, current_epoch=epoch+1, total_epochs=num_epochs, scaler=scaler)
        history['train_loss'].append(epoch_train_loss)
        history['train_acc'].append(epoch_train_acc)

        # 验证阶段
        epoch_val_loss, epoch_val_acc, _, _ = train_eval_one_epoch(model, dataloaders['val'], criterion, None, device, is_training=False, current_epoch=epoch+1, total_epochs=num_epochs)
        history['val_loss'].append(epoch_val_loss)
        history['val_acc'].append(epoch_val_acc)

        print(f"Epoch {epoch+1}/{num_epochs} Results -> Train Loss: {epoch_train_loss:.4f} Acc: {epoch_train_acc:.4f} | Val Loss: {epoch_val_loss:.4f} Acc: {epoch_val_acc:.4f}")
        if epoch_val_acc > best_val_acc_observed:
            best_val_acc_observed = epoch_val_acc # 更新观察到的最佳验证准确率

        # 早停判断和模型保存
        early_stopping(epoch_val_loss, model)

        # **** 清理CUDA缓存 ****
        if torch.cuda.is_available():
            torch.cuda.empty_cache()

        if early_stopping.early_stop:
            print("Early stopping triggered!")
            break

    time_elapsed = time.time() - since
    print(f'\nTraining complete in {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'Best validation accuracy (observed during training): {best_val_acc_observed:.4f}')

    return model, history

# --- 评估函数 (单GPU修改) ---
def evaluate_model_final(model, dataloader, criterion, device, class_names):
    model.eval()
    running_loss = 0.0
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for inputs, labels in tqdm(dataloader, desc="Evaluating Model", leave=False):
            inputs = inputs.to(device)
            labels = labels.to(device)
            # **** 混合精度：评估时也使用 autocast ****
            with autocast(enabled=torch.cuda.is_available()):
                outputs = model(inputs)
                final_outputs = outputs.logits if isinstance(outputs, models.inception.InceptionOutputs) else outputs
                loss = criterion(final_outputs, labels)
                _, preds = torch.max(final_outputs, 1)
            running_loss += loss.item() * inputs.size(0)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    total_loss = running_loss / len(dataloader.dataset)
    accuracy = np.mean(np.array(all_preds) == np.array(all_labels))
    cm = sm.confusion_matrix(all_labels, all_preds)
    return {'loss': total_loss, 'accuracy': accuracy, 'confusion_matrix': cm,
            'predictions': all_preds, 'true_labels': all_labels}

def plot_confusion_matrix(cm, class_names, filename='confusion_matrix.png', figsize=(12, 10), max_classes_to_display=20):
    num_classes = len(class_names)
    display_cm = cm
    display_class_names = class_names

    effective_num_classes = min(num_classes, max_classes_to_display)
    fig_width = max(12, effective_num_classes * 0.4)
    fig_height = max(10, effective_num_classes * 0.4)
    figsize = (fig_width, fig_height)

    if num_classes > max_classes_to_display and max_classes_to_display > 1:
        print(f"Number of classes ({num_classes}) exceeds max display ({max_classes_to_display}). Aggregating others.")
        sums_per_class = cm.sum(axis=1)
        sorted_indices = np.argsort(sums_per_class)[::-1]
        top_indices = sorted_indices[:max_classes_to_display - 1]
        other_indices = sorted_indices[max_classes_to_display - 1:]

        display_cm = np.zeros((max_classes_to_display, max_classes_to_display), dtype=cm.dtype)
        for i, r_idx in enumerate(top_indices):
            for j, c_idx in enumerate(top_indices):
                display_cm[i, j] = cm[r_idx, c_idx]
            display_cm[i, max_classes_to_display - 1] = cm[r_idx, other_indices].sum()

        for j, c_idx in enumerate(top_indices):
            display_cm[max_classes_to_display - 1, j] = cm[other_indices, c_idx].sum()
        display_cm[max_classes_to_display - 1, max_classes_to_display - 1] = cm[
            np.ix_(other_indices, other_indices)].sum()
        
        display_class_names = [class_names[i] for i in top_indices] + ["Other"]

    plt.figure(figsize=figsize)
    cm_normalized = display_cm.astype('float') / (display_cm.sum(axis=1)[:, np.newaxis] + 1e-9)
    
    sns.heatmap(cm_normalized, annot=True, fmt='.2f', cmap='Blues', xticklabels=display_class_names,
                yticklabels=display_class_names,
                linewidths=0.5, cbar_kws={'label': 'Normalized Prediction Ratio'},
                annot_kws={"size": 6})

    plt.title('Normalized Confusion Matrix', fontsize=16, fontweight='bold')
    plt.ylabel('True Label', fontsize=14)
    plt.xlabel('Predicted Label', fontsize=14)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(rotation=0, fontsize=10)
    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    print(f"混淆矩阵已保存为 {filename}")


# --- 绘制训练历史曲线 (保持不变) ---
def plot_training_history(history, filename='training_validation_curves.png'):
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history['train_loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(history['train_acc'], label='Training Accuracy')
    plt.plot(history['val_acc'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.savefig(filename, dpi=300)
    print(f"训练/验证曲线图已保存为 {filename}")


# **** 主训练函数 (单GPU修改) ****
def main(args: dict):
    # 设备设置
    device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
    print(f"正在使用设备: {device}")
    if str(device) == "cpu":
        print("警告: 未检测到CUDA GPU，将回退到CPU。训练速度会非常慢。")
        print("在Windows上，DataLoader的num_workers > 0 可能导致卡住。如果遇到问题，请尝试将num_workers设置为0。")


    train_dir = args['train_dir']
    test_dir = args['test_dir']

    if not os.path.exists(train_dir) or not os.path.exists(test_dir):
        print(f"错误: 数据目录未找到。请确认您的数据集路径是否正确。")
        exit()

    # --- 3. 数据加载 ---
    full_train_dataset_for_split = datasets.ImageFolder(root=train_dir)

    num_train_val_samples = len(full_train_dataset_for_split)
    train_size = int(0.8 * num_train_val_samples)
    val_size = num_train_val_samples - train_size

    generator = torch.Generator().manual_seed(args['seed']) # 确保划分一致

    all_indices = list(range(num_train_val_samples))
    train_subset_indices_list, val_subset_indices_list = random_split(all_indices, [train_size, val_size],
                                                                      generator=generator)

    train_subset_pil = Subset(full_train_dataset_for_split, train_subset_indices_list)
    val_subset_pil = Subset(full_train_dataset_for_split, val_subset_indices_list)

    train_dataset = TransformedSubset(train_subset_pil, train_transforms)
    val_dataset = TransformedSubset(val_subset_pil, val_test_transforms)

    test_dataset = datasets.ImageFolder(root=test_dir, transform=val_test_transforms)

    # 获取类别名称和数量
    class_names = full_train_dataset_for_split.classes
    NUM_CLASSES = len(class_names)
    print(f"数据集中的类别数量: {NUM_CLASSES}")
    print(f"总的训练/验证数据池大小: {len(full_train_dataset_for_split)}")
    print(f"划分后训练集大小: {len(train_dataset)}")
    print(f"划分后验证集大小: {len(val_dataset)}")
    print(f"独立测试集大小: {len(test_dataset)}")
    print("-" * 30)


    final_train_dl = DataLoader(train_dataset, batch_size=args['batch_size'], shuffle=True,
                                num_workers=args['num_workers'], pin_memory=True)
    final_val_dl = DataLoader(val_dataset, batch_size=args['batch_size'], shuffle=False,
                              num_workers=args['num_workers'], pin_memory=True)

    dataloaders = {'train': final_train_dl, 'val': final_val_dl}


    model_final = get_model(NUM_CLASSES, pretrained=True).to(device)


    if args['optimizer_name'] == "AdamW":
        optimizer_final = optim.AdamW(model_final.parameters(), lr=args['lr'], weight_decay=args['weight_decay'])
    else:
        optimizer_final = optim.SGD(model_final.parameters(), lr=args['lr'], momentum=0.9, weight_decay=args['weight_decay'])


    criterion_final = nn.CrossEntropyLoss()

    print("\n--- 开始使用固定超参数进行最终训练 ---")
    print(f"固定参数: LR={args['lr']:.6f}, Optimizer={args['optimizer_name']}, Batch Size={args['batch_size']}")
    print(f"Weight Decay: {args['weight_decay']}, Label Smoothing: Disabled") # 更新打印信息
    print(f"Number of Epochs: {args['num_epochs']}, Early Stopping Patience: {args['early_stopping_patience']}")
    print(f"Automatic Mixed Precision (AMP): {'Enabled' if torch.cuda.is_available() else 'Disabled (CUDA not available)'}")



    trained_model, history = train_loop(model_final, criterion_final, optimizer_final,
                                             dataloaders,
                                             num_epochs=args['num_epochs'],
                                             early_stopping_patience=args['early_stopping_patience'],
                                             checkpoint_path=args['checkpoint_path'],
                                             device=device)

    print("\n--- 在独立测试集上评估最佳模型 ---")

    model_for_eval = get_model(NUM_CLASSES, pretrained=True).to(device)
    if os.path.exists(args['checkpoint_path']):
        print(f"Loading best model saved by early stopping: {args['checkpoint_path']}")
        model_for_eval.load_state_dict(torch.load(args['checkpoint_path'], map_location=device))
    else:
        print(f"Warning: Checkpoint path {args['checkpoint_path']} not found. Using final state of trained model.")
        model_for_eval.load_state_dict(trained_model.state_dict())



    final_test_dl_eval = DataLoader(test_dataset, batch_size=args['batch_size'],
                                    shuffle=False, num_workers=args['num_workers'],
                                    pin_memory=True)

    test_results = evaluate_model_final(model_for_eval, final_test_dl_eval, criterion_final, device, class_names)
    print(f"测试集 | Loss: {test_results['loss']:.4f} Acc: {test_results['accuracy']:.4f}")


    print("\n--- 绘制训练/验证曲线图 ---")
    plot_training_history(history, filename='training_validation_curves_single_gpu.png')


    print("\n--- 绘制混淆矩阵 ---")
    plot_confusion_matrix(test_results['confusion_matrix'], class_names,
                          filename='confusion_matrix_single_gpu.png', max_classes_to_display=25)


    final_model_path = args['final_model_path']
    torch.save({
        'epoch_completed': args['num_epochs'],
        'model_state_dict': model_for_eval.state_dict(),
        'optimizer_state_dict': optimizer_final.state_dict(),
        'test_loss': test_results['loss'],
        'test_accuracy': test_results['accuracy'],
        'class_names': class_names,
        'training_history': history,
        'fixed_lr': args['lr'],
        'fixed_optimizer_name': args['optimizer_name'],
        'fixed_batch_size': args['batch_size'],
        'num_epochs_trained': len(history['train_loss']),
        'early_stopping_patience': args['early_stopping_patience'],
        'weight_decay': args['weight_decay'],
        'label_smoothing': "Disabled"
    }, final_model_path)
    print(f"\n最终模型和训练信息已保存到: {final_model_path}")


if __name__ == "__main__":
    # 路径配置
    script_train_dir = SPECIFIC_TRAIN_DIR
    script_test_dir = SPECIFIC_TEST_DIR

    if not os.path.exists(script_train_dir) or not os.path.exists(script_test_dir):
        print(f"错误: 数据集路径未找到。请检查路径。")
        exit()

    # 超参数字典 (调整为单GPU参数)
    args_dict = {
        'train_dir': script_train_dir,
        'test_dir': script_test_dir,
        'batch_size': 64, # 批次大小为64
        'num_workers': 0, # 保持为 0
        'optimizer_name': "AdamW",
        'lr': 0.001,
        'weight_decay': 1e-4,
        'label_smoothing': 0, # 此处值不再直接用于损失函数，仅作为记录
        'num_epochs': 30,
        'early_stopping_patience': 7,
        'checkpoint_path': 'single_gpu_inceptionv3_best_model.pth',
        'final_model_path': 'single_gpu_inceptionv3_final_model.pth',
        'seed': 42
    }

    print(f"Starting single GPU training with InceptionV3.")
    print(f"Batch Size: {args_dict['batch_size']}")
    print(f"Number of Workers for DataLoader: {args_dict['num_workers']}")
    print(f"Learning rate: {args_dict['lr']:.6f}")
    print(f"Weight Decay: {args_dict['weight_decay']}")
    print(f"Label Smoothing: Disabled (value was {args_dict['label_smoothing']})")
    print(f"Number of Epochs: {args_dict['num_epochs']}")
    print(f"Early Stopping Patience: {args_dict['early_stopping_patience']}")
    print(f"Checkpoint Path: {args_dict['checkpoint_path']}")
    print(f"Final Model Path: {args_dict['final_model_path']}")

    main(args_dict)
    print("Single GPU training finished.")
```
